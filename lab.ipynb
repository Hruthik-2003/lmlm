## iii. Plot the histogram for continuous variables (at least two) to analyse the data.

```python
import matplotlib.pyplot as plt
import numpy as np
# Plot histogram for a continuous variable
plt.figure(figsize=(12, 6))

# Histogram for the first continuous variable
plt.subplot(1, 2, 1)
plt.hist(df['rating'].dropna(), bins=30, edgecolor='black')
plt.title('Histogram of Rating')
plt.xlabel('Value')
plt.ylabel('Frequency')
bins = np.linspace(min(df['price'].dropna()), 100000, 30)
# Histogram for the second continuous variable
plt.subplot(1, 2, 2)
plt.hist(df['price'].dropna(), bins=bins, edgecolor='black',color="#f1856f")
plt.title('Histogram of Price')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## iv. Draw a violin plot do describe the distribution of a numerical variable to analyse the data.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x=df['processor_speed'].dropna(), data=df)

# Customize the plot,
plt.title('Violin Plot of Values by Category')
plt.xlabel('processor_speed in GHz')
plt.ylabel('Value')

plt.show()
```

```python

# Function to plot box plots
def plot_boxplot(data, column, title):
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=data[column])
    plt.title(title)
    plt.show()

# Function to treat outliers (you can customize this part)
def treat_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(.75)
    IQR = Q3 - Q1
#     IQR proximity rule
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Removing outliers
    data_treated = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    return data_treated

# Column to analyze (replace 'Price' with your column of interest)
column = 'price'

# Plot box plot before outlier treatment
plot_boxplot(df, column, 'Before Outlier Treatment')

# Treat outliers
df_treated = treat_outliers(df, column)

# Plot box plot after outlier treatment
plot_boxplot(df_treated, column, 'After Outlier Treatment')

# Save the treated dataset (optional)
# df_treated.to_csv('mobile_phones_treated.csv', index=False)
```
